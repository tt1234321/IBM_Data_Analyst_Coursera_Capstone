## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
```

```python
df['ConvertedComp'].plot(kind='box', figsize=(10, 8))
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
```

```python
df[['ConvertedComp']].describe()
```

```python
Qu1=df['ConvertedComp'].quantile(0.25)
Qu3=df['ConvertedComp'].quantile(0.75)
IQR=Qu3-Qu1
print("interquantile range=Q3(75%)-Q1(25%)=", Qu3, " - ", Qu1, "=", IQR)
```

Find out the upper and lower bounds.


```python
# your code goes here
```

```python
lower=Qu1-1.5*IQR
lower
```

```python
upper=Qu3+1.5*IQR
upper
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
df[(df['ConvertedComp'] < lower) | (df['ConvertedComp'] > upper)].ConvertedComp.count()
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
```

```python
df_rem_out=np.clip(df[['ConvertedComp']], lower, upper, out=None)
df_rem_out
```

```python
df_rem_out.mean()
```

```python
df_rem_out.median()
```

```python
df_rem_out.describe()
```

```python
df_rem_out.plot(kind='box', figsize=(10, 8))
```

```python
#### or I will first remove missing values and than repeat steps
```

```python
df_no_miss_val=df.ConvertedComp.dropna()
df_no_miss_val
```

```python
df_no_miss_val.isnull().value_counts()
```

```python
df_no_miss_val_Qu1=df_no_miss_val.quantile(0.25)
df_no_miss_val_Qu3=df_no_miss_val.quantile(0.75)
df_no_miss_valIQR=df_no_miss_val_Qu3-df_no_miss_val_Qu1
print("interquantile range=Q3(75%)-Q1(25%)=", df_no_miss_val_Qu3, " - ", df_no_miss_val_Qu1, "=", df_no_miss_valIQR)
```

```python
new_lower=df_no_miss_val_Qu1-1.5*df_no_miss_valIQR
new_lower
```

```python
new_upper=df_no_miss_val_Qu3+1.5*df_no_miss_valIQR
new_upper
```

```python
df_new_remove_outliers=np.clip(df_no_miss_val, new_lower, new_upper, out=None)
df_new_remove_outliers.describe()
```
