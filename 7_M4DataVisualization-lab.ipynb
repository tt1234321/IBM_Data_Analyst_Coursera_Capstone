```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

```

```python
Query = """
SELECT AVG(WorkWeekHrs), AVG(CodeRevHrs), Age 
FROM master 
where Age between 30 and 35
group by Age
"""

df_stack = pd.read_sql_query(Query, conn)
df_stack
```

```python
### or
```

```python
Query = """
SELECT WorkWeekHrs, CodeRevHrs, Age 
FROM master 
where Age between 30 and 35
"""

df_stack2 = pd.read_sql_query(Query, conn)
df_stack2
```

```python
df_stack_data_median=df_stack2.groupby(by=["Age"]).median()
df_stack_data_median
```

```python
df_stack_data_mean=df_stack2.groupby(by=["Age"]).mean()
df_stack_data_mean
```

```python

WorkWeekHrs= df_stack_data_mean['WorkWeekHrs']
CodeRevHrs= df_stack_data_mean['CodeRevHrs']
Age= df_stack_data_mean.index

fig, ax = plt.subplots()
figsize=(15, 6)
ax.bar(Age, WorkWeekHrs, label='WorkWeekHrs')
ax.bar(Age, CodeRevHrs, bottom=WorkWeekHrs,
       label='CodeRevHrs')

plt.show()
```

```python
WorkWeekHrs= df_stack_data_mean['WorkWeekHrs']
CodeRevHrs= df_stack_data_mean['CodeRevHrs']
Age= df_stack_data_mean.index

fig, ax = plt.subplots(figsize=(15, 10))

ax.bar(Age, WorkWeekHrs, label='WorkWeekHrs')
ax.bar(Age, CodeRevHrs, bottom=WorkWeekHrs,
       label='CodeRevHrs')

plt.title('Stacked chart of MEAN WorkWeekHrs and CodeRevHrs for the age group 30 to 35') # add a title to the histogram
plt.ylabel('WorkWeekHrs plus CodeRevHrs (MEAN)') # add y-label
plt.xlabel('Age') # add x-label


plt.show()
```

```python
WorkWeekHrs= df_stack_data_median['WorkWeekHrs']
CodeRevHrs= df_stack_data_median['CodeRevHrs']
Age= df_stack_data_median.index

fig, ax = plt.subplots(figsize=(15, 10))

ax.bar(Age, WorkWeekHrs, label='WorkWeekHrs')
ax.bar(Age, CodeRevHrs, bottom=WorkWeekHrs,
       label='CodeRevHrs')

plt.title('Stacked chart of median WorkWeekHrs and CodeRevHrs for the age group 30 to 35') # add a title to the histogram
plt.ylabel('WorkWeekHrs plus CodeRevHrs (MEDIAN)') # add y-label
plt.xlabel('Age') # add x-label


plt.show()
```
